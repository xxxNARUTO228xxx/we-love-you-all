# Визуальный анализ кластеров

## Подготовка данных

```python
drop_string_cols = ["image_path", "image_viz_path", "class_name"]
df_nums = df.drop(drop_string_cols, axis=1)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df_nums)
```

```python
df_nums_scaled = scaler.transform(df_nums)
```

## PCA

```python
# https://machinelearningknowledge.ai/complete-tutorial-for-pca-in-python-sklearn-with-example/
pca2 = PCA(n_components=2)
principalComponents = pca2.fit_transform(df_nums_scaled)

principalDf = pd.DataFrame(data = principalComponents, columns = ['pca_1', 'pca_2'])

finalDf = pd.concat([principalDf, df[['class_name']]], axis = 1)
finalDf.head()
```

```python
plt.figure(figsize=(12,7))
 
sns.scatterplot(data=finalDf, 
                x="pca_1", 
                y="pca_2", 
                hue="class_name")
 
plt.title("Figure 1: Scatter Plot",
          fontsize=16)
plt.xlabel('Principal Component 1',
           fontsize=16)
plt.ylabel('Principal Component 2',
           fontsize=16)
```

## TSNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42, verbose=1)
X_tsne = tsne.fit_transform(df_nums.to_numpy())
```

```python
tnse_df = pd.DataFrame(data = X_tsne, columns = ['tnse_1', 'tnse_2'])

tnse_df = pd.concat([tnse_df, df[['class_name']]], axis = 1)
tnse_df.head()
```

```python
plt.figure(figsize=(12,7))
 
sns.scatterplot(data=tnse_df, 
                x="tnse_1", 
                y="tnse_2", 
                hue="class_name")
 
plt.title("Figure 1: Scatter Plot",
          fontsize=16)
plt.xlabel('TSNE Component 1',
           fontsize=16)
plt.ylabel('TSNE Component 2',
           fontsize=16)
```
