SPLIT TASK ON 5 PERSON

```python
import os
import shutil
from random import shuffle

# –ü—É—Ç—å –∫ –ø–∞–ø–∫–µ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏
source_folder = ''

# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞–ø–æ–∫ –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è
n_folders = 5

# –ü–æ–ª—É—á–∏—Ç—å —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ –ø–∞–ø–∫–µ
image_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

# –ü–µ—Ä–µ–º–µ—à–∞—Ç—å —Å–ø–∏—Å–æ–∫ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
shuffle(image_files)

# –†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –ø–æ –ø–∞–ø–∫–∞–º
images_per_folder = len(image_files) // n_folders
remainder = len(image_files) % n_folders

# –°–æ–∑–¥–∞—Ç—å N –ø–∞–ø–æ–∫ –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
for i in range(n_folders):
    # –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∏–º—è –Ω–æ–≤–æ–π –ø–∞–ø–∫–∏
    new_folder_name = os.path.join(source_folder, f'person_{i+1}')
    
    # –°–æ–∑–¥–∞—Ç—å –Ω–æ–≤—É—é –ø–∞–ø–∫—É, –µ—Å–ª–∏ –æ–Ω–∞ –µ—â–µ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
    os.makedirs(new_folder_name, exist_ok=True)
    
    # –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å, —Å–∫–æ–ª—å–∫–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Å–∫–æ–ø–∏—Ä–æ–≤–∞–Ω–æ –≤ —ç—Ç—É –ø–∞–ø–∫—É
    images_to_copy = image_files[i*images_per_folder : (i+1)*images_per_folder + (1 if i < remainder else 0)]
    
    # –ö–æ–ø–∏—Ä–æ–≤–∞—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ –Ω–æ–≤—É—é –ø–∞–ø–∫—É
    for img in images_to_copy:
        shutil.copy2(os.path.join(source_folder, img), new_folder_name)

print("–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –±—ã–ª–∏ —É—Å–ø–µ—à–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã.")
```
